# 2.进阶绘图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 不同形状和颜色绘图
t = np.arange(0, 5, 0.2)  # 建立numpy数组[0, 5), d=0.2
# red dashes, blue squares, and green triangles
plt.plot(t, t, "r--", t, t**2, "bs", t, t**3, "g^")  # 连续在同一图上绘图(x1,y1,x2,y2)
plt.show()
```

```python
# 不同形状和颜色绘图
x = np.linspace(0, 2, 101)
# Note that even in the 00-style, we use `.pyplot.figure` to create
fig, ax = plt.subplots()  # create a figure and an axes.
ax.plot(x, x, label="linear")  # plot some data on the axes.
ax.plot(x, x**2, label="quadratic")  # plot more data on the axes.
ax.plot(x, x**3, label="cubic")  # ... and some more.
ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Simple Plot")
ax.legend()  # set a legend
plt.show()
```

```python
x = np.linspace(0, 10, 500)
y = np.sin(x)
fig, ax = plt.subplots()
# Using set_dashes() to modify dashing of an existing line
line1, = ax.plot(x,y,label="Using set_dashes")
line1.set_dashes([2, 2, 10, 2])  # 2pt line, 2pt break, 10pt line 
line2, = ax.plot(x, y-0.2, dashes=[6,2],label="Using the dasehs parameter")
ax.legend()
plt.show()
```
