# 1.基本绘图

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 使用matplotlib创建简易图
plt.plot([1,2,3,4])  # 默认为y轴序列，x轴默认从0开始，即plt.plot(y)
plt.show()
```

```python
# 添加x轴序列
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])  # plt.plot(x, y)
plt.show()
```

```python
# 添加x轴和y轴标签
plt.rcParams['font.sans-serif'] = ['Songti SC']
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.ylabel("因变量")
plt.xlabel("自变量")
plt.show()
```

```python
# 画数据点
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "ro--")
plt.show()
```

```python
# 调整x，y轴坐标范围
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], "ko-")
plt.axis([0, 6, 0, 20])
plt.show()
```
